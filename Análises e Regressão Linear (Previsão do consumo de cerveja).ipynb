## Iniciando Modelagem de Machine Learning

#### Estimar o valor médio do consumo referente as variáveis expicativas


```python
#importando o train test split
from sklearn.model_selection import train_test_split
```

```python
#criando uma serie
y = dados['consumo']
```

```python
#criando um dataframe com as variáveis explicativas que vão ser utilizadas no modelo
X = dados[['temp_max','chuva','fds']]
```

```python
#separando os x e y - treino e teste - Random state(Gerador aleátorio que permite que os resultados de treino e teste sempre sejamos mesmos)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2811)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Função de regressão linear multopla com 3 variáveis explicativas
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
modelo = LinearRegression()
```

```python
#Treinando o modelo
modelo.fit(X_train, y_train)
```

```python
#Calculando o Coeficiente de Determinação - O quanto a linha de regressão ajusta aos dados - Quanto mais proximo de 1 melhor
modelo.score(X_train, y_train).round(2)
```

## Gerando previsões para o dado de teste

```python
y_previsto = modelo.predict(X_test)
```

```python
#obter o r2 para o y teste
metrics.r2_score(y_test, y_previsto).round(2)
```

```python
#GRÁFICO PARA VERIFICAR O AJUSTE DOS RESÍDUOS - SUA DISPOSIÇÃO NO GRÁFICO
#QUANTO MAIS O RESIDUO CONTORNAR O EIXO PROXIMO DE 0 MELHOR É O AJUSTE - HOMOCEDASTICIDADE DOS DADOS
ax = sns.scatterplot(x=y_previsto_train, y=residuo, s=150)
ax.figure.set_size_inches(20, 8)
ax.set_title('Resíduos X Previsão', fontsize=18)
ax.set_xlabel('Consumo de Cerveja (litros) - Previsão', fontsize=14)
ax.set_ylabel('Resíduos', fontsize=14)
ax
```

```python
#residuos elevados ao quadrado e verificar o formato do gráfico
ax = sns.scatterplot(x=y_previsto_train, y=residuo**2, s=150)
ax.figure.set_size_inches(20, 8)
ax.set_title('Resíduos X Previsão', fontsize=18)
ax.set_xlabel('Consumo de Cerveja (litros) - Previsão', fontsize=14)
ax.set_ylabel('Resíduos²', fontsize=14)
ax
```
